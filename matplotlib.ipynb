```python
import matplotlib.pyplot as plt
```

```python
# Çizgi Grafiği
x = [ 1, 2, 3, 4, 5 ]
y = [ 100, 200, 300, 400, 500 ]

plt.plot(x,y, marker="8", linestyle="dashed", color="red")
plt.grid()
plt.xlabel("Günler")
plt.ylabel("Satışlar")
plt.show()
```

```python
# Çizgi Grafikleri
```

```python
import numpy as np

x = np.linspace(0,10,100) # 0 ile 10 arasında eşit aralıklarda 100 sayı oluştur.
```

```python
# X değerinin sinüs ve cos. çizgi grafiğini tek grafikte göster.
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# plt.plot() -> direkt kullanılırsa tekil bir çizgi grafiği verir

plt.figure(figsize=(8,6)) # yeni boş bi kağıt gibi düşünelim.
plt.plot(x,y1, label="sin(x)", color="red")
plt.plot(x,y2, label="cos(x)", color="blue")
plt.legend() # -> Grafiğin bilgi alanını aktif et
plt.show()
```

## Bar Grafiği

```python
categories = ["A","B","C","D","E","F"]
values = [5,10,15,20,25,30]
```

```python
plt.bar(categories,values, color=["red","green","blue"])
plt.xlabel("Kategoriler")
plt.ylabel("Satış Değerleri")
plt.title("Kategori Bazlı Satış Grafiği")
plt.show()
```
